```python
# Function to initialize the inventory
def initialize_inventory(products):
    inventory = {}
    for product in products:
        quantity = int(input(f"Enter the quantity for {product}: "))
        inventory[product] = quantity
    return inventory

# Function to get customer orders
def get_customer_orders(products):
    customer_orders = set()
    while True:
        order = input("Enter the name of the product to order: ")
        if order in products:
            customer_orders.add(order)
        else:
            print("The product is not available in the list.")
        
        another = input("Do you want to add another product? (yes/no): ").lower()
        if another != "yes":
            break
    return customer_orders

# Function to update inventory based on customer orders
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if inventory[product] > 0:
            inventory[product] -= 1
        else:
            print(f"Sorry, {product} is out of stock!")
    return inventory

# Function to calculate order statistics
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    total_available_products = len(products)
    percentage_ordered = (total_products_ordered / total_available_products) * 100
    return total_products_ordered, percentage_ordered

# Function to print order statistics
def print_order_statistics(order_statistics):
    total_products_ordered, percentage_ordered = order_statistics
    print("Order Statistics:")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Products Ordered: {percentage_ordered:.2f}%")

# Function to print updated inventory
def print_updated_inventory(inventory):
    print("Updated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

# Main
def main():
    # 1: Define list of products
    products = ["t-shirt", "mug", "hat", "book", "keychain"]
    
    # 2: Initialize inventory
    inventory = initialize_inventory(products)
    
    # 3: Get customer orders
    customer_orders = get_customer_orders(products)
    
    # 4: Calculate and print order statistics
    order_statistics = calculate_order_statistics(customer_orders, products)
    print_order_statistics(order_statistics)
    
    # 5: Update inventory based on customer orders
    updated_inventory = update_inventory(customer_orders, inventory)
    
    # 6: Print updated inventory
    print_updated_inventory(updated_inventory)

# Boot it!
if __name__ == "__main__":
    main()
```
